```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image_read = cv2.imread("card_2.jpg")
```

```python
if image_read is not None:
    plt.imshow(image_read)
    plt.show()
else:
    print("Error loading the image. Please check the file path.")
```

```python
image_color = cv2.cvtColor(image_read, cv2.COLOR_BGR2RGB)
image_color
```

```python
plt.imshow(image_color)
```

```python
image_hsv = cv2.cvtColor(image_read, cv2.COLOR_RGB2HSV)
image_hsv
```

```python
plt.imshow(image_hsv)
```
